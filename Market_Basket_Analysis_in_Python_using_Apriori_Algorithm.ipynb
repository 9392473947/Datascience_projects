Implementing market basket analysis

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df=pd.read_excel('http://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx')
```

```python
df
```

```python
df.head()
```

Data Preparation

```python
df['Description']=df['Description'].str.strip() #removing spaces from beginign and ending
```

```python
#removing duplicates
df.dropna(axis=0,subset=['InvoiceNo'],inplace=True)
```

```python
#converting invoice to str
df['InvoiceNo']=df['InvoiceNo'].astype('str')
```

```python
#removing credit transactions
df=df[~df['InvoiceNo'].str.contains('C')]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Country'].value_counts()
```

```python
df.shape
```

```python
#separating transaction for germany
mybasket=(df[df['Country']=="Germany"]
          .groupby(['InvoiceNo','Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
mybasket.head()
```

```python
#converting all positive values to 1 and everything else to 0
def my_encode_units(x):
  if x<=0:
    return 0
  if x>=1:
    return 1
mybasket_sets = mybasket.applymap(my_encode_units)
mybasket_sets.drop('POSTAGE',inplace=True,axis=1) #remove postage as an item
```

```python
mybasket_sets
```

Traning model

```python
#generating frequent itemsets
my_frequent_itemsets = apriori(mybasket_sets,min_support=0.07,use_colnames=True)
my_frequent_itemsets
```

```python
my_rules = association_rules(my_frequent_itemsets,metric='lift',min_threshold=1)
```

```python
my_rules.head()
```

Making recommendations

```python
#filtering rules based on condition
my_rules[(my_rules['lift']>=3) & (my_rules['confidence']>=0.3)]
```
