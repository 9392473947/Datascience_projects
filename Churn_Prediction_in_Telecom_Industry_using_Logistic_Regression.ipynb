Exploratory Data Analysys

```python
df.describe()
```

```python
df['TotalCharges'].describe()
```

```python
#total charges have a string at random place we need to convert int
#so convert string to null and then drop null values so that it convert to int
df['TotalCharges']=df['TotalCharges'].replace(" ",np.nan)
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
df=df.dropna(how='any',axis=0)
```

```python
df
```

```python
df['TotalCharges'].describe()
```

```python
df.notnull().sum()
```

```python
df['Churn']=df['Churn'].replace(['Yes','No'],[1,0])
df
```

```python
df_getdummies=pd.get_dummies(df)
df_getdummies
```

```python
df_getdummies.corr()
```

```python
df_getdummies['Churn'].sort_values(ascending=False)
```

```python
sns.heatmap(df_getdummies.corr())
```

Train test split

```python
x=df_getdummies.drop(['Churn'],axis=1)
```

```python
x
```

```python
y=df_getdummies['Churn']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(x_train,y_train)
```

```python
y_pred=lg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
